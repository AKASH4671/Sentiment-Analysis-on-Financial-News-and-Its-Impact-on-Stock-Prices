**libraries**

```python
import yfinance as yf
import pandas as pd
```

**get stock data**

```python
# Define the date range from your news data
start_date = "2024-06-24"
end_date = "2025-07-22"

# Fetch Apple (AAPL) stock price data
ticker = yf.Ticker("AAPL")
df_prices = ticker.history(start=start_date, end=end_date)

# Reset index to get 'Date' column
df_prices.reset_index(inplace=True)
```

**save**

```python
# Save to Drive
output_path = "/content/drive/MyDrive/Colab Notebooks/Finance Projects/Sentiment-Analysis-on-Financial-News-and-Its-Impact-on-Stock-Prices/data/apple_prices.csv"
df_prices.to_csv(output_path, index=False)

print(f"Apple price data saved to:\n{output_path}")
print("\nDate range in stock price data:")
print("Start:", df_prices['Date'].min())
print("End:  ", df_prices['Date'].max())

# Preview
df_prices.head()
```
