**Libraries**

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

**Fetch Stock Price**

```python
# Define date range based on your scraped news
start_date = "2024-10-01"
end_date = "2025-07-21"

# Fetch Apple stock price from Yahoo Finance
ticker = yf.Ticker("AAPL")
df_prices = ticker.history(start=start_date, end=end_date)

# Reset index to get date column
df_prices.reset_index(inplace=True)

# Save to Drive
output_path = "/content/drive/MyDrive/Colab Notebooks/Finance Projects/Sentiment-Analysis-on-Financial-News-and-Its-Impact-on-Stock-Prices/data/apple_prices.csv"
df_prices.to_csv(output_path, index=False)

print(f" Saved Apple prices to {output_path}")
```

**Plot**

```python
# Set style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(14, 6))
plt.plot(df_prices['Date'], df_prices['Close'], color='blue', linewidth=2)

plt.title("Apple (AAPL) Daily Close Price", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Close Price (USD)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
